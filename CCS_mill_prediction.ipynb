```python
data = pd.read_excel('NIR_brix_mill.xlsx')
data.shape
```

```python
# Convert 'Brix(F)' column to numeric
data['Brix(F)'] = pd.to_numeric(data['Brix(F)'], errors='coerce')

# Check the data types
print(data.dtypes)
```

```python
import numpy as np

# Calculate z-scores for each column
z_scores = (data - data.mean()) / data.std()

# Set the threshold for outlier detection (e.g., z-score > 3 or < -3)
threshold = 3

# Identify outliers based on the z-scores
outliers = (z_scores > threshold) | (z_scores < -threshold)

# Replace outliers with NaN or remove them from the dataset
data[outliers] = np.nan

# Handle the outliers based on your analysis (e.g., imputation or removal)

# Print the modified dataset
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
data.info()
```

```python
data
```

```python
data=data[data['Mill']==9]
```

```python
data.info()
```

```python
data.rename(columns = {'Brix(F)':'BrixF'}, inplace = True)
data.rename(columns = {'%Fiber':'Fiber'}, inplace = True)
```

```python
X = data[['BrixF', 'Fiber']]    
y = data['CCS'] 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
import xgboost as xgb
import matplotlib.pyplot as plt
from xgboost import plot_importance, plot_tree
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
%%time
parameters = {
    'n_estimators': [100,200,400,500],
    'learning_rate': [0.001,0.01],
    'max_depth': [10,15,20,30],
    'gamma': [0.001, 0.005, 0.01],
    'random_state': [42]
}
#X_trainv, X_testv, y_trainv, y_testv = train_test_split(X_train, y_train, test_size=0.15, random_state=42)
eval_set = [(X_train, y_train),(X_test, y_test)]
model = xgb.XGBRegressor(eval_set=eval_set, objective='reg:squarederror', verbose=3)
clf = GridSearchCV(model, parameters,cv=5)

clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
predictions = clf.predict(X_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('R^2', metrics.r2_score(y_test, predictions))
```
